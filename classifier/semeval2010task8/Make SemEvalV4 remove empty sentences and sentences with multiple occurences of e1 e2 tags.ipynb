```python
print(len(data))

# returns true if a label if contained twice in a sentence.
# there is nothing wrong with these data items themselves but they
# were processed incorrectly in an earlier stage and need to be removed.
def duplicated(sent, label):
    index = sent.find(label)
    if index == -1:
        return False
    second = sent.find(label, index)
    print(sent, label)
    return second == -1

```

```python
filtered = []
for d in data:
    if d["e1_contents"] != "" and d["e2_contents"] != "":
        filtered.append(d)

print(len(filtered))


with open("semeval_testV4.json", "w") as f:
    json.dump(filtered, f, indent = 4)
```
